# ---Bar Chart---

```python
import matplotlib.pyplot as plt
course = ["Python","Java","C#"]
score = [80,65,50]
```

```python
plt.bar(course,score)
plt.xlabel("Course", color = "blue", backgroundcolor = "pink")
plt.ylabel("Score", color = "blue", backgroundcolor = "pink")
plt.title("Score Midterm", color = "red", backgroundcolor = "yellow")
plt.show()
```

```python
c = ["green","orange","red"]
plt.bar(course, score, color=c, width=0.25, edgecolor="blue", linewidth=10)
plt.xlabel("Course", color = "blue", backgroundcolor = "pink")
plt.ylabel("Score", color = "blue", backgroundcolor = "pink")
plt.title("Score Midterm", color = "red", backgroundcolor = "yellow")
plt.show()
```

```python
c = ["green","orange","red"]
plt.barh(course, score, color=c, height=0.5, edgecolor="magenta", linewidth=5)
plt.xlabel("Score", color = "blue", backgroundcolor = "pink")
plt.ylabel("Course", color = "blue", backgroundcolor = "pink")
plt.title("Score Midterm", color = "red", backgroundcolor = "yellow")
plt.show()
```

# Stackbar / Stackplot

```python
# 3   Stackbar
room = ["A","B","C"]
boys = [10,15,5]
girls = [20,10,15]
plt.bar(room,boys,label="Men")
plt.bar(room,girls,bottom=boys,label="Women",color="pink")
# bottom=boys boysgirls bottom 
plt.legend()
plt.show()
```

```python
plt.stackplot(room, boys, girls, labels=["Men", "Women"],colors=["cyan", "pink"])
plt.legend()
plt.show()
```
