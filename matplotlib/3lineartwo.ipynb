```python
import matplotlib.pyplot as plt
product_1 = [10, 10, 20, 30, 50, 80, 140]    # ยอดขายสินค้า1
product_2 = [15, 30, 45, 60, 30, 60, 90]    # ยอดขายสินค้า2
month = [1, 2, 3, 4, 5, 6, 7]    # เดือนที่...
```

plt.title("สรุปผลข้อมูล", loc = "left")
plt.title("สรุปผลข้อมูล", loc = "right")

```python
plt.plot(month, product_2, "g+--")
data = {"size" : 20, "color" : "red", "backgroundcolor" : "yellow"}
plt.xlabel("Month", data)
plt.ylabel("Sales", data)
config = {"size" : 20, "color" :"blue"}
plt.title("Top Sale 2020", config, loc="right")
plt.show()
```

# Face & Edge Color Legend
- facecolor สีของแผ่น legend
    plt.legend(facecolor="red")
    plt.legend(facecolor="yellow")
- edgecolor สีของเส้น legend
    plt.legend(edgecolor="red")
    plt.legend(facecolor="yellow", edgecolor="red")

```python
plt.plot(month, product_1, "g+-")
plt.plot(month, product_2, "r*-")
data = {"size" : 20, "color" : "red", "backgroundcolor" : "yellow"}
plt.xlabel("Month", data)
plt.ylabel("Sales", data)
config = {"size" : 20, "color" :"blue"}
plt.title("Top Sale 2020", config, loc="right")
plt.legend(["Mouse", "Keyboard"], facecolor="cyan", edgecolor="red")
plt.show()
```

# Title Legend
- title เป็นการกำหนดข้อความ มีค่าเริ่มต้นคือNone
    plt.legend(title="ข้อความ")
    plt.legend(title=None)
    plt.legend(title="ข้อความ", title_fontsize="small")
    plt.legend(borderpad=3)     // ขนาด default(ขนาดเริ่มต้น) = 0.4
    plt.legend(labelspacing=3)  // ระยะ default(ระยะเริ่มต้น) = 0.5

```python
plt.plot(month, product_1, "g+-")
plt.plot(month, product_2, "r*-")
data = {"size" : 20, "color" : "red", "backgroundcolor" : "yellow"}
plt.xlabel("Month", data)
plt.ylabel("Sales", data)
config = {"size" : 20, "color" :"blue"}
plt.title("Top Sale 2020", config, loc="center")
plt.legend(["Mouse", "Keyboard"],
title="Product",
title_fontsize="xx-large",
fontsize="large",
facecolor="cyan", 
edgecolor="black",
borderpad=3,
labelspacing=0.5)
plt.show()
```

# Text Function

```python
plt.plot(month, product_1, "g+-")
plt.plot(month, product_2, "r*-")
data = {"size" : 20, "color" : "red", "backgroundcolor" : "yellow"}
plt.xlabel("Month", data)
plt.ylabel("Sales", data)
config = {"size" : 20, "color" :"blue"}
plt.title("Top Sale 2020", config)
plt.text(4,63,"point 1",size=20,color="blue")
plt.text(5,25,"point 2",size="small",color="magenta")
plt.show()
```
