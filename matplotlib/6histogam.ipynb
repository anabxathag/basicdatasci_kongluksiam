# ---Histogram---
สำหรับแสดงความถี่ของข้อมูล เช่น แสดงความถี่ของอายุลูกค้าที่ใช้ระบบงานของเรา
(สำหรับข้อมูลไม่ต่อเนื่อง)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
age = [18,17,20,18,20,15,19,20,17,16,15,14,16,17,20]
plt.hist(age)
plt.xlabel("Age", color = "blue", backgroundcolor = "pink")
plt.ylabel("Count", color = "blue", backgroundcolor = "pink")
plt.title("Total Age", color = "red", backgroundcolor = "yellow")
plt.show()
```

# histogram (สำหรับข้อมูลต่อเนื่อง)
สุ่มข้อมูล5000ชุด มีค่าเฉลี่ย10 และส่วนบี่ยงเบน2.5

```python
data = np.random.normal(10,2.5,5000)
plt.hist(data)      # ไม่กำหนดช่วงข้อมูล (bins=10)
plt.show()
```

```python
plt.hist(data, bins=30)      # กำหนดช่วงเป็น30กลุ่ม
plt.show()
```
