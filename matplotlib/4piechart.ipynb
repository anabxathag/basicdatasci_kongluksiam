# ---กราฟวงกลม(Pie Charts)---
สำหรับแสดงข้อมูลที่มีสัดส่วนรวมกันได้ 100% หรือเน้นส่วนที่สนใจเป็นพิเศษ
data = [15,30,45,60,40];
lang=['PHP','Flutter','Nodejs','C#','Java'];
colors=['green','orange','yellow','red','cyan']
plt.pie(data,labels=lang)
autopct='%.1f%%'    // ใส่ค่า%ในแต่ละตัว

```python
import matplotlib.pyplot as plt
student = [10,20,40,30,60]
course = ['PHP','Flutter','Nodejs','C#','Java']
```

```python
plt.pie(student, labels=course, autopct='%.2f%%')
plt.show()
```

```python
colo=['green','orange','yellow','red','cyan']
exp = [0,0.2,0.4,0.6,0.3]
plt.pie(student, labels=course, autopct='%.1f%%',colors=colo,shadow=True,explode=exp,startangle=90)
plt.show()
```
