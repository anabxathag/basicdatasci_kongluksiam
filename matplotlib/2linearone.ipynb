# ---กราฟเส้น---

```python
import matplotlib.pyplot as plt
product1 = [10, 10, 20, 30, 50, 80, 140]    # ยอดขายสินค้า1
product2 = [15, 30, 45, 60, 30, 60, 90]    # ยอดขายสินค้า2
month = [1, 2, 3, 4, 5, 6, 7]    # เดือนที่...
```

```python
plt.plot(month, product1)
plt.plot(month, product2)
plt.xlabel("Month")     # คำอธิบายแกนX
plt.ylabel("Sales")     # คำอธิบายแกนY
plt.savefig("plotingwhite.png")     # พื้นหลังขาว
plt.savefig("plotingcrystal.png",transparent=True)     # พื้นหลังโปร่งแสง
plt.show()      # แสดงกราฟ
```

# Color Support (รูปแบบสี)
'b' = blue
'g' = green
'r' = red
'c' = cyan
'm' = magenta
'y' = yellow
'k' = black
'w' = white

```python
plt.plot(month, product1, color = "m")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

# Marker Support (รูปแบบจุด)
'.' = point
'o' = circle
'x' = x marker
'+' = plus
'*' = star
's' = square
'|' = vline
'_' = hline
'v' = down
'^' = up
'>' = right
'<' = left

```python
plt.plot(month, product1, "*")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

```python
plt.plot(month, product1, color='r', marker='x')
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

# Line Support (รูปแบบเส้น)
'-' = solid
'--' = dashed
'-.' = dash-dot
':' = dotted
# กราฟเส้น(Format String)
    [market][line][color]
    [color][market][line]
- การใช้งาน
    plt.plot([10,20,30],[1,2,3,4],'r+-')

```python
plt.plot(month, product1, color='g', marker='o', ls='-.')
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

```python
plt.plot(month, product1, "ms:")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

# กำหนดคุณสมบัติของLabel
// คุณสมบัติผ่าน dict
data = {"size" : 20, "color" : "black"}
plt.xlabel("ข้อความ", data)
// คุณสมบัติผ่าน properties
plt.xlabel("ข้อความ", size = 20, color = "black", backgroundcolor = "red")

```python
plt.plot(month, product1, "g>--")
plt.xlabel("Month", size = 20, color = "red", backgroundcolor = "yellow")
plt.ylabel("Sales", size = 24, color = "blue", backgroundcolor = "pink")
plt.show()
```

```python
plt.plot(month, product1, "y^--")
data = {"size" : 20, "color" : "blue", "backgroundcolor" : "pink"}
plt.xlabel("Month", data)
plt.ylabel("Sales", data)
plt.show()
```
